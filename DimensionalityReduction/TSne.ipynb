```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

appleData = pd.read_csv('apple_quality.csv') #predict if apple has good or bad quality
appleData = appleData.iloc[:1000] 
data = appleData.iloc[:, 1:appleData.shape[1]-1]

scaler = StandardScaler()
data = scaler.fit_transform(data)
                            
labels = appleData['Quality'].replace({'good': 1, 'bad': 0}) #encode labels

xtrain, xtest, ytrain, ytest = train_test_split(data, labels, test_size=0.2, random_state=42)
```

```python
p = 25
tsne = TSNE(n_components=2, random_state=42, perplexity=p)
xtrain_tsne = tsne.fit_transform(xtrain)
```

```python
plt.figure(figsize=(5, 4))
plt.scatter(xtrain_tsne[ytrain==1, 0], xtrain_tsne[ytrain==1, 1], label='1', c='blue', alpha=0.5)
plt.scatter(xtrain_tsne[ytrain==0, 0], xtrain_tsne[ytrain==0, 1], label='0', c='red', alpha=0.5)

plt.title(f'perplexity: {p}')
plt.xlabel('t-SNE feature 1')
plt.ylabel('t-SNE feature 2')
plt.legend()
plt.show()
```
